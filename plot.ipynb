```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('./result/time1.csv')
```

```python
torch_data = data[data['type']=='torch']
float_data = data[data['type']=='float']
int_data = data[data['type']=='int']
```

```python
torch_conv_data = torch_data[torch_data['target']=='conv']
float_conv_data = float_data[float_data['target']=='conv']
int_conv_data = int_data[int_data['target']=='conv']

torch_linear_data = torch_data[torch_data['target']=='linear']
float_linear_data = float_data[float_data['target']=='linear']
int_linear_data = int_data[int_data['target']=='linear']

torch_all_data = torch_data[torch_data['target']=='all']
float_all_data = float_data[float_data['target']=='all']
int_all_data = int_data[int_data['target']=='all']
```

```python
torch_conv_data_1 = torch_conv_data[torch_conv_data['batch']==1]
torch_conv_data_64 = torch_conv_data[torch_conv_data['batch']==64]
float_conv_data_1 = float_conv_data[float_conv_data['batch']==1]
float_conv_data_64 = float_conv_data[float_conv_data['batch']==64]
int_conv_data_1 = int_conv_data[int_conv_data['batch']==1]
int_conv_data_64 = int_conv_data[int_conv_data['batch']==64]

torch_linear_data_1 = torch_linear_data[torch_linear_data['batch']==1]
torch_linear_data_64 = torch_linear_data[torch_linear_data['batch']==64]
float_linear_data_1 = float_linear_data[float_linear_data['batch']==1]
float_linear_data_64 = float_linear_data[float_linear_data['batch']==64]
int_linear_data_1 = int_linear_data[int_linear_data['batch']==1]
int_linear_data_64 = int_linear_data[int_linear_data['batch']==64]

torch_all_data_1 = torch_all_data[torch_all_data['batch']==1]
torch_all_data_64 = torch_all_data[torch_all_data['batch']==64]
float_all_data_1 = float_all_data[float_all_data['batch']==1]
float_all_data_64 = float_all_data[float_all_data['batch']==64]
int_all_data_1 = int_all_data[int_all_data['batch']==1]
int_all_data_64 = int_all_data[int_all_data['batch']==64]
```
